# limpa colunas

```python
import unidecode

def clean_column_names(df):
    """
    Limpa os nomes das colunas de um DataFrame para garantir que sejam compatíveis com nomes de colunas em bancos de dados.
    - Remove espaços e caracteres especiais.
    - Converte caracteres acentuados para não acentuados.
    - Garante que o nome não comece com números.
    """
    new_columns = []
    for column in df.columns:
        # Remove caracteres especiais e substitui espaços por underscore
        clean_column = unidecode.unidecode(column).replace(" ", "_")
        clean_column = ''.join(char for char in clean_column if char.isalnum() or char == '_')
        
        # Se a coluna começa com número, adiciona um prefixo
        if clean_column[0].isdigit():
            clean_column = 'col_' + clean_column
        
        new_columns.append(clean_column.lower())  # Opcional: converter para minúsculo

    df.columns = new_columns
    return df


dff = clean_column_names(dff)
dff
```

# BigQuery

```python
from google.cloud import bigquery

class BigQuery:
    def __init__(self, frame, project_id, target_table  ) -> None:
        self.__project_id = project_id
        self.__client = bigquery.Client(project=self.__project_id)
        self.__target_table = target_table
        self.__frame = frame
        try:
            print(self.__frame.info())
        except Exception as e:
            print('erro em obter informações')
        self.__job_config = bigquery.LoadJobConfig(
            autodetect=True,
            write_disposition = 'WRITE_TRUNCATE' #'WRITE_APPEND',
        )

    def upload(self):
        load_job = self.__client.load_table_from_dataframe(
            self.__frame,
            self.__target_table,
            job_config=self.__job_config
        )
        print('Carregando tabela no BigQuery...')
        load_job.result()
        table = self.__client.get_table(self.__target_table)
        print(table.num_rows, '\n')
```

```python
from google.cloud import bigquery
from google.oauth2 import service_account
import pandas as pd

class BigQuery:
    def __init__(self, frame, project_id, dataset_id, table_id, credentials_path='databridgex-prototipo-c1a4d51704c8.json') -> None:
        self.__project_id = project_id
        # Carrega as credenciais do arquivo JSON especificado
        self.__credentials = service_account.Credentials.from_service_account_file(credentials_path)
        # Cria o cliente do BigQuery com as credenciais fornecidas
        self.__client = bigquery.Client(project=self.__project_id, credentials=self.__credentials)
        # Monta o identificador completo da tabela no formato "project.dataset.table"
        self.__target_table = f"{self.__project_id}.{dataset_id}.{table_id}"
        self.__frame = frame
        try:
            print(self.__frame.info())
        except Exception as e:
            print(f'Erro em obter informações: {str(e)}')
        self.__job_config = bigquery.LoadJobConfig(
            autodetect=True,
            write_disposition='WRITE_APPEND'  # Pode ser mudado para WRITE_APPEND se necessário
        )

    def upload(self):
        load_job = self.__client.load_table_from_dataframe(
            self.__frame,
            self.__target_table,
            job_config=self.__job_config
        )
        print('Carregando tabela no BigQuery...')
        load_job.result()  # Aguarda a conclusão do job
        table = self.__client.get_table(self.__target_table)
        print(f'Carregado {table.num_rows} linhas para a tabela {self.__target_table}')

```
